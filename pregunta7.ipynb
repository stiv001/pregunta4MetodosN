# Splines Cubicos 

```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_spline(x, x_i, coeficientes):
    dx = x - x_i
    return coeficientes["a"] + coeficientes["b"] * dx + coeficientes["c"] * dx**2 + coeficientes["d"] * dx**3

def graficar_tangente(x1, y1, m):
    x = np.array([x1-0.5, x1+0.5])
    y = m * (x - x1) + y1
    plt.plot(x, y, '--b', label='Tangente')

xs = np.array([-1, 0, 1])
ys = np.array([1, 5, 3])

splines = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = calcular_spline(_x, x_i, splines[i])
    plt.plot(_x, _y, color="red")

graficar_tangente(0, 5, 2)

plt.scatter(xs, ys)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con Splines Cúbicos y Tangente")
plt.legend()
plt.grid(True)

plt.show()
```
